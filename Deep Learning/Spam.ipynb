```python
# Select only columns that start with "word_freq_"
word_freq_columns = [col for col in df.columns if col.startswith('word_freq_')]

# Rename selected columns for 'word_freq_' columns
df.rename(columns=lambda x: x.split('_')[-1] if x in word_freq_columns else x, inplace=True)

# Map symbols for both 'word_freq_' and 'capital_run_length_' columns
df.columns = df.columns.map(lambda x: x.split('_')[-1]).map(lambda x: {'leftbrac': '{', 'leftsquarebrac': '[', 'exclaim': '!', 'dollar': '$', 'pound': '£'}.get(x, x))

# Display the DataFrame with renamed columns
print(df)
```

```python
from sklearn.model_selection import train_test_split
features = df_filter.columns[:-1]
X = df_filter[features]
y = df_filter['spam']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## DNN

```python
from keras.models import Sequential
from keras.layers import Dense
#from keras.utils import to_categorical

# Build the DNN model
model = Sequential([
          Dense(100, input_dim=X_train_scaled.shape[1], activation='relu', name='input_layer'),
          Dense(64, activation='relu', name='hidden_layer_1'),
          Dense(32, activation='relu', name='hidden_layer_2'),
          Dense(1, activation='sigmoid', name='output_layer')
])
```

```python
model.summary()
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred_binary)
print("Classification Report:")
print(report)
```

```python
#add regularization to model to avoid overfitting
from keras import regularizers

# Define the regularization strength (you can adjust this value)
l2_strength = 0.01

# Build the DNN model with L2 regularization
model_reg = Sequential([
    Dense(100, input_dim=X_train_scaled.shape[1], activation='relu', kernel_regularizer=regularizers.l2(l2_strength), name='input_layer'),
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2_strength), name='hidden_layer_1'),
    Dense(32, activation='relu', kernel_regularizer=regularizers.l2(l2_strength), name='hidden_layer_2'),
    Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(l2_strength), name='output_layer')
])
```

```python
model_reg.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Train the model on the standardized training set
model_reg.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

```python
y_pred = model_reg.predict(X_test_scaled)
y_pred_binary = (y_pred >= 0.5).astype(float)

ytrain = model_reg.predict(X_train_scaled)
y_train_binary = (ytrain >= 0.5).astype(float)
```

```python
# Evaluate the accuracy of the model
accuracy_test = accuracy_score(y_test, y_pred_binary)
accuracy_train = accuracy_score(y_train, y_train_binary)

print(f"Accuracy train: {accuracy_train:.2f}")
print(f"Accuracy test: {accuracy_test:.2f}")
```

```python
report = classification_report(y_test, y_pred_binary)
print("Classification Report:")
print(report)
```
